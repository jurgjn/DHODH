```python
def DiffDockCsv(ligand_name, protein_name, protein_path, ligand_description, protein_sequence, path):
    df_ = pd.DataFrame()
    df_['complex_name'] = ligand_name + f'+{protein_name}'
    df_['protein_path'] = protein_path
    df_['ligand_description'] = [* map(smi_largest_component, ligand_description) ]
    df_['protein_sequence'] = protein_sequence
    df_.to_csv(path, index=False, header=True, sep=',')
```

```python
def read_picked(fp_):
    with open(fp_) as fh_:
        lig_ = fh_.read(7)
        assert lig_ == 'ligand\t'
        df_ = pd.read_csv(fh_, sep='\t', names=['smiles', 'name', 'protonation_id'])
        df_['compound_id'] = [f'{r["name"]}_{r.protonation_id}' for i, r in df_.iterrows() ]
        #df_['compound_id'] = df_.at[0, 'compound_id']
        #df_['is_decoy'] = True
        #df_.at[0, 'is_decoy'] = False
        return df_

for i, r in df_.iterrows():
    print(r.compounds_id, r.picked_file)
    df_picked = read_picked(r.picked_file)
    DiffDockCsv(df_picked['compound_id'], '1d3g', os.path.abspath('results/pdb/1d3g.pdb'), df_picked['smiles'], '', f'results/DiffDockCsv/{r.compounds_id}+1d3g.csv')
    DiffDockCsv(df_picked['compound_id'], '1d3h', os.path.abspath('results/pdb/1d3h.pdb'), df_picked['smiles'], '', f'results/DiffDockCsv/{r.compounds_id}+1d3h.csv')
    DiffDockCsv(df_picked['compound_id'], '4igh', os.path.abspath('results/pdb/4igh.pdb'), df_picked['smiles'], '', f'results/DiffDockCsv/{r.compounds_id}+4igh.csv')
    DiffDockCsv(df_picked['compound_id'], '6j3c', os.path.abspath('results/pdb/6j3c.pdb'), df_picked['smiles'], '', f'results/DiffDockCsv/{r.compounds_id}+6j3c.csv')
```
